```python
import pandas as pd
monthly_progress4=[]
task = ["Presentation","Reading","Assignment","Disscussion"]
total = 0
for i in task:
    time1 = input ("Enter hrs spent on task "+i +" in 1st_week:")
    time2 = input ("Enter hrs spent on task "+i +" in 2nd_week:")
    time3 = input ("Enter hrs spent on task "+i +" in 3rd_week:")
    time4 = input ("Enter hrs spent on task "+i +" in 4th_week:")
    
    monthly_progress4.append({"Task":i,"1st_Week":time1,"2nd_Week":time2,"3rd_Week":time3,"4th_Week":time4})
print (monthly_progress4)
df = pd.DataFrame(monthly_progress4)
df.to_csv("monthly_progress4.csv",mode="w",index=False)
df
```

```python
import pandas as pd
from matplotlib import pyplot as plt
df = pd.read_csv("monthly_progress4.csv")
df
time_list= df.values.tolist()
print("total time list: ",time_list)
task=[]
week1=[]
week2=[]
week3=[]
week4=[]
for i in time_list:
    task.append(i[0])
    week1.append(i[1])
    week2.append(i[2])
    week3.append(i[3])
    week4.append(i[4])

tw1=0
for i in week1:
    tw1+= i
    avg1=tw1/4
tw2=0
for i in week2:
    tw2+= i 
    avg2=tw2/4
tw3=0
for i in week3:
    tw3+= i
    avg3=tw3/4
tw4=0
for i in week4:
    tw4+= i
    avg4=tw4/4
total_week =[tw1,tw2,tw3,tw4]
avg=[avg1,avg2,avg3,avg4]
print ("total hrs per week:",total_week)
print ("Average:",avg)

tp= time_list[0]
tr= time_list[1]
ta= time_list[2]
td= time_list[3]
task1=tp[1:5]
task2=tr[1:5]
task3=ta[1:5]
task4=td[1:5]
week=[1,2,3,4]
print ("task1",task1)
print ("task2",task2)
print ("task3",task3)
print ("task4",task4)
print ("week",week)
monthly_progress4.append({"Task": "Total hrs per week",'1st_Week': tw1, '2nd_Week': tw2, '3rd_Week': tw3, '4th_Week': tw4})

df = pd.DataFrame.from_dict(monthly_progress4)
df 
```
